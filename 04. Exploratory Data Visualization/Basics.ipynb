### Plotting using Pandas

#### Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs Sample_size', figsize=(10,10))
```

```python
ax = recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', figsize=(8,5))
ax.set_title('Employed vs Sample_size Scatter')
```

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs Sample_size Scatter', figsize=(8, 5))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs Sample_size Scatter', figsize=(8, 5))
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs Full_time Scatter', figsize=(8, 5))
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs ShareWomen Scatter', figsize=(8, 5))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Median vs Men Scatter', figsize=(8, 5))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Median vs Women Scatter', figsize=(8, 5))
```
